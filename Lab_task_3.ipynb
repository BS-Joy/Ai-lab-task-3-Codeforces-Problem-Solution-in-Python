```python
courses = []
choice = "random"


def lines1():
    print('------------------------')


def lines2():
    print('**************************')


def menu():
    lines2()
    print('Menu')
    print('''
    1. Add Course\n
    2. Show Course\n
    3. remove Course\n
    4. Exit''')
    lines1()
    choice = input('Enter Your Choice:')
    return choice


while choice != '4':
    choice = menu()
    if choice == '1':
        course = input("Enter Course Name/Code:")
        courses.append(course)
        print(f"{course} is added to your course list.")
        print()
    elif choice == '2':
        if len(courses) == 0:
            print("Sorry! You haven't add any course to the list yet.")
            print()
        else:
            print('List of your Courses:')
            for course in courses:
                print(course)
    elif choice == '3':
        if len(courses) == 0:
            print("Sorry! You haven't add any course to the list yet.")
            print()
        else:
            course = input('Enter course name/code to remove:')
            if course in courses:
                courses.remove(course)
                print(f'{course} is removed from the course list')
            else:
                print(f'{course} is not found to your course list')
    elif choice == '4':
        print('Goodbye, have a great day.')
    else:
        print('Wrong input! Please input between 1, 2, 3 and 4. Thank You!')
```
